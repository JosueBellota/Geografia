# Ejemplo de integración GPX (XML) y Dataframe

```python
path = os.getcwd()
# input(string): solicita al usuario la introducción de un valor
gpx = input("Introduce nombre archivo GPX (sin extensión):")
pathgpx = path + '/' + gpx + ".gpx"
# Creamos el XML y Dataframe
docxml = et.parse(pathgpx)
raiz = docxml.getroot()

waypoints = range (1,5,1) # ejemplo para 5 puntos
columnas = ['Punto','Latitud','Longitud', 'Diferencia Longitud', 'Diferencia Latitud']
df = pd.DataFrame(columns=columnas)

#-----------------------------------------------------------------

def diferencias (lat1,lat2,lon1,lon2):
    dla = (lat2-lat1)
    dlo = (lon2 -lon1)
    
    return (dlo,dla) 

# -----------------------------------------------------------------

for i in waypoints:
    # Datos del punto A
    wpt1 = raiz[i]
    lat1 = float(wpt1.get("lat"))
    lon1 = float(wpt1.get("lon"))
    
    
    if i == 1:
        N = 0.00
        df_1 = pd.DataFrame([[i,lat1,lon1,N,N]],columns = list(columnas))
        df = pd.concat([df if not df.empty else None,df_1],ignore_index=True)  #controla si un dataframe esta vacio   
        
# Incrementamos punto       
    i = i+1

# Datos del punto B
    wpt2 = raiz[i]
    lat2 = float(wpt2.get("lat"))
    lon2 = float(wpt2.get("lon"))
    
    
# Llamamos a las diferentes funciones
    dlo,dla = diferencias(lat1,lat2,lon1,lon2)
    df_1 = pd.DataFrame([[i,lat2,lon2,dlo,dla]],columns = list(columnas))
    df = pd.concat([df,df_1],ignore_index=True)
    
# --------------------------------------------------------------------    

df.head(5)
```
